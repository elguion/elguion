```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi ,100)
y = np.cos(x)

plt.plot(x ,y)
plt.title ("Cos graph")
plt.xlabel( " x-values")
plt.ylabel( "Cosine values")


```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace (0 , 2*np.pi , 30 , dtype = int)
y = np.sin(x)


plt.plot(x ,y)


plt.title(" Sine Graph")
plt.xlabel("X-valuses")
plt.ylabel(" Sin x")
plt.grid(True)
```
